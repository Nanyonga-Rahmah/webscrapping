#No.1 Comparing series using pandas
we use the compare() method , greater than, less than , and equal-to operators 
after importing the neccesary libraries
the compare method prints the difference between the series that is the index
positions at which the elements did not match


```python
import pandas as pd
series1=pd.Series([4,65,436,3,9])
series2=pd.Series([7,0,3,897,9])
print("Greater Than \n", series1>series2)
print("Less Than \n ", series1<series2)
print("Equal To \n ", series1==series2)
print("Difference \n ", series1.compare(series2))
```

#No.2Program to add,subtract,divide and multiply
we use the arithmetic operators that is + - * /


```python
serie1=pd.Series([2,4,6,8,14])
serie2=pd.Series([1,3,5,7,9])
print("Adding\n", serie1 + serie2)
print("Subtracting\n", serie1 - serie2)
print("Multiplying\n", serie1 * serie2)
print("Dividing\n", serie1 / serie2)
```
